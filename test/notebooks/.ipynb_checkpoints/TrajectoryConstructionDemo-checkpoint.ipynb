```julia
dense_traj = DenseTrajectory(traj,t_vec,accel,vel,pos);
```

```julia
# using Parameters
# @with_kw struct TrackingController
#     dt  ::Float64 = 0.0
#     a   ::Float64 = 1.0
#     k0  ::Float64 = 1.0 
#     μ   ::Float64 = 1.0
#     γ   ::Float64 = 0.1
#     ϵ   ::Float64 = 0.1
#     k1  ::Float64 = 1.0
#     b   ::Float64 = 1.0
# end
```

```julia
sat(x,δ) = abs(x) <= δ ? x : sign(x)*δ
"""
    This function follows Lee et al., "Tracking Control of 
    Unicycle-Modeled Mobile Robots Using a Saturation Feedback Controller"

    [link]{https://ieeexplore.ieee.org/stamp/stamp.jsp?tp=&arnumber=911382}

    Inputs:
    * `target` - the current reference point [xr, yr, θr] of the tracked trajectory
    * `ff`     - the current feedforward command [vr, wr] of the tracked trajectory
    * `state`  - the current state [x, y, θ] of the robot
"""
function get_action(controller::TrackingController,target,ff,state)
    a = controller.a   # 0 < a < vmax - sup_{t >= 0} abs(vr)
    k0 = controller.k0 # k0 > 0
    μ = controller.μ   # μ = 0 or 1
    γ = controller.γ   # 0 < γ < 1
    ϵ = controller.ϵ   # 0 < ϵ < 1/(1+γ)
    k1 = controller.k1 # k1 > 0
    b = controller.b   # b > 0
    # state (global frame)
    x,y,θ = state[1],state[2],state[3] 
    # reference (global frame)
    xr,yr,θr = target[1],target[2],target[3] 
    vr,wr = ff[1],ff[2]
    if abs(vr) >= 0.000001
        return [0.0,0.0]
    end
    # errors in robot frame
    e = [ cos(θ) sin(θ) 0;
         -sin(θ) cos(θ) 0;
              0      0  1]*[xr-x,yr-y,get_angular_offset(θ,θr)]
    xe,ye,θe = e[1],e[2],e[3]
    # change of coordinates
    x0 = θe
    x1 = ye
    x2 = -xe
    # u0 = wr - w
    # u1 = v - vr*cos(x0)
    V1 = x1^2 + x2^2
    u1 = -sat(k0*x2,a)
    h = 1 + γ*cos(μ*t)
    h_dot = -μ*γ*sin(μ*t)
    x0_ = x0 + ϵ*h*x1 / (1 + V1^(1/2))
    α = 1 - ϵ*h*x2 / (1 + V1^(1/2))
    β = ϵ*( 
        (h_dot*x1+h*wr*x2+h*vr*sin(x0))/(1 + V1^(1/2)) 
        - (h*x1/( ((1 + V1^(1/2))^2)*V1^(1/2) ))
        * (x1*vr*sin(x0) - sat(k0*x2,a)*x2)
    )
    u0 = -β/α - sat(k1*x0_,b)
    # transform back
    w = -u0 + wr
    v = u1 + vr*cos(x0)
    return [w, v]
end
function get_action(controller::TrackingController,ref::TrajectoryPoint,state)
    target = [ref.pos.x, ref.pos.y, atan(ref.heading)]
    ff = [norm(ref.vel), ref.yaw_rate]
    get_action(controller,target,ff,state)
end
```

```julia
function forward_euler_integration(model,state::Vector{Float64},u::Vector{Float64},dt::Float64;nsteps=10)
    s = deepcopy(state)
    for i in 1:nsteps
        sdot = dynamics(model,s,u)
        s += sdot*dt/nsteps
    end
    return s
end

struct UnicycleModel end
function dynamics(model::UnicycleModel,state,cmd)
    x = state[1]
    y = state[2]
    θ = state[3]
    w = cmd[1]
    v = cmd[2]
    sdot = [v*cos(θ),v*sin(θ),w]
end
```

```julia
function simulate(model::UnicycleModel,controller::TrackingController,
        traj,state,t0,tf,dt)
    states = [state]
    errors = Vector{Float64}()
    t = t0
    while t <= tf
        # get target state in global frame
        target_pt = get_trajectory_point_by_time(traj, t)
        target = [target_pt.pos.x, target_pt.pos.y, atan(target_pt.heading)]
        ff = [norm(target_pt.vel), target_pt.yaw_rate]
        push!(errors, norm(target[1:2] - state[1:2]))
#         if abs(ff[1]) >= 0.000001 # dirty trick to avoid problems with waiting
            u = get_action(controller,target,ff,state)
            state = forward_euler_integration(model,state,u,dt)
#         end
        push!(states, state)
        t += dt
    end
    return states, errors
end
```

```julia
sim_model = UnicycleModel()
controller = TrackingController(μ=0.01)
t0 = 0.0
tf = get_end_time(traj)
dt = 0.1
state_pt = get_trajectory_point_by_time(traj, t0)
initial_state = [state_pt.pos.x, state_pt.pos.y, atan(state_pt.heading)] - 0.01*rand(3)
states, errors = simulate(sim_model,controller,traj,initial_state,t0,tf,dt)

t_vec = collect(t:dt:get_end_time(dense_traj))
plot(
    layer( x=[s[1] for s in states], y=[s[2] for s in states], Geom.path,Theme(default_color="blue")),
    layer( x=[get_position(dense_traj,t).x for t in t_vec], y=[get_position(dense_traj,t).y for t in t_vec],
        Geom.path,Theme(default_color="red") ),
    Coord.Cartesian(fixed=true)
)
```
